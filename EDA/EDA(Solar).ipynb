-Solar radiation shows a clear yearly seasonal pattern, peaking in summer and dropping in monsoon.

-Cloud cover has a strong negative impact on GHI, meaning more clouds  less solar radiation.

-Temperature increases with higher GHI, indicating sunny days produce higher temperatures.

-Humidity  and PW reduce GHI, showing moisture blocks sunlight.

-Aerosol  slightly decreases GHI, meaning dust and pollution reduce solar output.

-GHI distribution is almost normal, centered between 46, with few extreme outliers.

-Summer months MarJun achieve the highest GHI, while monsoon months show the lowest.

-Zenith Angle and Day Length strongly correlate with GHI, confirming that sun position controls radiation levels.
